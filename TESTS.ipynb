```python
p = []
m = []
a = 0
for i in ev:
    a+=1
    year1 = node_info[node_info["Id"]==i[0]]["Pubyear"].values[0]
    year2 = node_info[node_info["Id"]==i[1]]["Pubyear"].values[0]
    delta = year1-year2
    if delta > 0:
        p.append(delta)
    if delta < 0:
        m.append(delta)
    if a > 10000:
        break
```
